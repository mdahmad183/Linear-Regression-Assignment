```python
season_name = pd.get_dummies(day_csv['season'],drop_first = True)
month = pd.get_dummies(day_csv['mnth'],drop_first = True)
day_of_week = pd.get_dummies(day_csv['weekday'],drop_first = True)
weather_type = pd.get_dummies(day_csv['weathersit'],drop_first = True)
```

```python
day_csv = pd.concat([day_csv ,season_name], axis = 1)
day_csv = pd.concat([day_csv ,month], axis = 1)
day_csv = pd.concat([day_csv ,day_of_week], axis = 1)
day_csv = pd.concat([day_csv ,weather_type], axis = 1)
```
